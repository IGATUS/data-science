# FILTERING

```python
df2=df.loc[df['MILES*']>10,['MILES*']]
df2
```

```python
df[df['MILES*']>10].head()
```

+ When doing filtering of data, you cannot mutate the columns, whatever condition you use will be applied to the entire columns
+ so it is advised to use loc or iloc when doing filtering to achieve your purpose

```python
# find out all rides from Fort Pierce
df.loc[df['START*']=='Fort Pierce',['MILES*']]
```

```python
# find out all the rides to cary and morrisville
myrides=df.loc[df['START*'].isin(['Cary','Morrisville'])]
myrides.head()
```

### The isin operator is used to carry out multiple column selection as shown above, you can also use regular expression

```python
myrides.loc[0:5,:]
```

```python
myrides.iloc[0:5,:]
```
