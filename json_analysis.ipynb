# working with json

```python
book={}
book['tom']={
    'name':'tom',
    'address':'anyigba state',
    'phone':23243432
}
book['igatus']={
    'name':'igatus',
    'address':'kogi state',
    'phone':23243532
}
import json
s=json.dumps(book) # the s in dumps stands for string, thus it mean dump as a string
s
```

```python
with open('book.txt','w') as f:
    f.write(s)
```
